# Evaluation on the test set

In this notebook, we evaluate the model created by Azure AutoML: the model was created using a trainning set, lets see how it performs on the testing set

```python
from azureml.core import Workspace, Experiment
from azureml.core.compute import ComputeTarget
from azureml.widgets import RunDetails
from azureml.train.estimator import Estimator
import azureml.train.automl
```

```python
# load ws
ws = Workspace.from_config()

# Get the dataset
data = ws.datasets.get('poker_ds')
```

```python
# get compute target and start it
cpu_cluster = ComputeTarget(workspace=ws, name='pc3')
cpu_cluster.start()
cpu_cluster.wait_for_completion(show_output=True)
```

## The Experimento

```python
# Create an estimator
estimator = Estimator(
    source_directory='scripts',
    entry_script='check_score.py',
    compute_target='pc3',
    conda_packages=['scikit-learn'],
    inputs=[data.as_named_input('poker')],
    )

# Submit
exp = Experiment(workspace=ws, name='PokerHand_AutoML_score')

# Run the experiment based on the estimator
run = exp.submit(config=estimator)
RunDetails(run).show()
run.wait_for_completion(show_output=True)
```

## Stopping the compute target

```python
for pc_name in ws.compute_targets:
    pc = ws.compute_targets[pc_name]
    pc.stop()

```
